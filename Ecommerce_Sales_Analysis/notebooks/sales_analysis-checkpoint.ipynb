## E-Commerce Sales & Profitability Analysis

This project analyzes retail sales data to identify revenue trends,
product profitability, regional performance, and the impact of discounts.

The insights from this analysis were used to build an interactive
Power BI dashboard for business decision making.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("C:\\Users\\balaj\\OneDrive\\Desktop\\Ecommerce_Sales_Analysis\\data\\SampleSuperstore.csv\\superstore_sales.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
sales_region = df.groupby('Region')['Sales'].sum()
print(sales_region)
```

```python
sales_region.plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

```python
sales_category = df.groupby('Category')['Sales'].sum()
print(sales_category)
```

```python
sales_category.plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```python
profit_category = df.groupby('Category')['Profit'].sum()
print(profit_category)
```

```python
profit_category.plot(kind='bar')
plt.title("Profit by Category")
plt.show()
```

```python
profit_subcategory = df.groupby('Sub-Category')['Profit'].sum().sort_values()

profit_subcategory.plot(kind='barh')
plt.title("Profit by Sub-Category")
plt.xlabel("Profit")
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

```python
profit_subcategory = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

print(profit_subcategory)
```

```python
profit_subcategory.plot(kind='bar', figsize=(10,5))
plt.title("Profit by Sub-Category")
plt.show()
```

```python
sales_segment = df.groupby('Segment')['Sales'].sum()

print(sales_segment)
```

```python
sales_segment.plot(kind='bar')
plt.title("Sales by Customer Segment")
plt.show()
```

```python
state_profit = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)

print(state_profit)
```

```python
state_profit.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 States by Profit")
plt.show()
```

```python
df[['Sales','Profit','Discount','Quantity']].corr()
```

```python
sns.heatmap(df[['Sales','Profit','Discount','Quantity']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
loss_products = df.groupby('Sub-Category')['Profit'].sum().sort_values()

print(loss_products.head(10))
```

```python
loss_products.head(10).plot(kind='bar')
plt.title("Top Loss-Making Product Categories")
plt.show()
```

```python
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)

print(state_sales.head(10))
```
